```python
import cv2 as cv
from time import sleep
import os 
import numpy as np
```

```python
os.chdir('D:\WorkStation\object\computer_vision\car_count')
video = cv.VideoCapture('video.mp4')
fourcc = cv.VideoWriter_fourcc('X','V','I','D')
out = cv.VideoWriter('out.avi',fourcc,20,(1280,720))
back_ground = cv.createBackgroundSubtractorMOG2(history=100,varThreshold=50)
detc = []
offset =6
count = 0
```

```python
def center(x,y,w,h):
    w =int(w/2)
    h = int(h/2)
    x = x+w
    y = y+h
    return x,y
```

```python
while(True):
    ret,frame = video.read()
    if ret is not True:
        break
    sleep(1/60)    
    gray = cv.cvtColor(frame,cv.COLOR_BGR2GRAY)
    blur = cv.GaussianBlur(gray,(3,3),5)
    kernel =back_ground.apply(blur)
    dilate = cv.dilate(kernel,np.ones((5,5)))
    contours,_ = cv.findContours(dilate,cv.RETR_TREE,cv.CHAIN_APPROX_SIMPLE)
    cv.line(frame,(25,550),(1200,550),(0,255,255),2)
    for (i,c) in enumerate(contours):
        (x,y,w,h) = cv.boundingRect(c)
        validar_contorno = (w >= 80) and (h >= 80)                 
        if not validar_contorno:                 
         continue 
        cv.rectangle(frame,(x,y),(x+w,y+h),(0,255,255),2)
        cent = center(x,y,w,h)
        print(cent)
        detc.append(cent)
        cv.circle(frame,cent,4,(0,0,255),-1)
        for (x,y) in detc:
            if(y<(550+offset)) and (y>(550-offset)):
                count+=1
                cv.line(frame,(25,550),(1200,550),(0,0,255),3)
                detc.remove((x, y))    

    cv.putText(frame, "VEHICLE COUNT : "+str(count), (320, 70),cv.FONT_HERSHEY_COMPLEX, 2, (0, 0, 255), 4)
    cv.imshow('mask',dilate)
    cv.imshow('Frame',frame)
    frame = cv.resize(frame, (1280,720))
    out.write(frame)
    key = cv.waitKey(30)
    if key ==27 :
        break
video.release()
cv.destroyAllWindows()
    
```
